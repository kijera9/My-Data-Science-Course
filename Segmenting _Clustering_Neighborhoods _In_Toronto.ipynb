We specify the url of the wikipedia page we want to scrape and used the urllib.request library to query the web page an dput the html date into the variable(url).

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
canada = urllib.request.urlopen(url)
```

Import BeautifulSoup library

```python
from bs4 import BeautifulSoup
```

Use BeautifulSoup to parse the html data from the url and store in a new variale call soup_canada

```python
soup_canada = BeautifulSoup(canada, "html.parser")
soup_canada.title
```

Get the instance of all table in the wiki page using find_all method and store in a new variable call all_tables.

```python
all_tables=soup_canada.find_all("table")
all_tables
```

Get the table with the class id wikitable sortable and store in right_table

```python
right_table=soup_canada.find('table', { "class" : "wikitable sortable" })
right_table
```

Create a list of A, B and C which are the three columns in the right table and use a for loop to iterate through the table table row(tr) using the findALl function, then assigned each table data(td) to a cell and finally find the text for each cell and append it to each cell of the table data in the each row and column.

```python
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
    
```

Import pandas dataframe and create a DataFrame (df) and assign each list A, B and C into a column and fill them with the text Postal Code, Borough and Neighborhood. Pick only Boroughs that are assigned and remove the special character \n which is all part of cleaning the data.

```python
import pandas as pd

df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighborhood']=C

only_assigned_borough = (df[df.Borough != 'Not assigned\n'].reset_index(drop=True)).replace('\n','', regex=True)

only_assigned_borough

```
